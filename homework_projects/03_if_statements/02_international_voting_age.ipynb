# **Problem Statement**
Write a program which asks a user for their age and lets them know if they can or can't vote in the following three fictional countries.

# ***SOlutuion:***

```python
def main():
  print("Welcome to program")
  user_age:int=int(input("Enter your age. "))

  if user_age >= 16:
    print("You can vote in Peturksbouipo where the voting age is 16 " )
  else:
    print("You can't vote in Peturksbouipo where the voting age is 16")

  if user_age >= 25:
    print("You can vote in Stanlau where the voting age is 25. ")
  else:
    print("You can't vote in Stanlau where the voting age is 25. ")

  if user_age >= 48:
    print("You can vote in Mayengua where the voting age is 48. ")
  else:
    print("You can't vote in Mayengua where the voting age is 48. ")

  print("Hope you get your information!")

if __name__ == '__main__':
    main()
```
