# **Problem Statement**
Guess My Number

# ***Solution:***

```python
import random
def main():
  Number:int= random.randint(1,99)
  print("I have Guessed a number between 0 to 99")
  Guessed_num=int(input("Guess which Number i guessed? "))
  while Guessed_num !=  Number:
    if Guessed_num < Number:
      print("Your Guess is incorrect.")
    else:
      print("Your Guess is too High")
    print()
    Guessed_num=int(input("Try again."))
  print("congrats! You guessed right number")


if __name__ == '__main__':
    main()

```
