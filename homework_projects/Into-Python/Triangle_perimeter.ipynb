# ***Problem Statement:***
Prompt the user to enter the lengths of each side of a triangle and then calculate and print the perimeter of the triangle (the sum of all of the side lengths).



# ***Solution:***

```python
def main():
 Len_1=float(input("Enter the length of side 1 "))
 Len_2=float(input("Enter the length of side 2 "))
 Len_3=float(input("Enter the length of side 3 "))
 F_Len=Len_1+Len_2+Len_3
 print(f"The perimeter of the triangle is: {F_Len} ")

if __name__ == '__main__':
    main()
```
