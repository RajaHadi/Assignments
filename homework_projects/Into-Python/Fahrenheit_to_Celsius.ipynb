# **Problem Statement**
Write a program which prompts the user for a temperature in Fahrenheit (this can be a number with decimal places!) and outputs the temperature converted to Celsius.

# **Solution**

```python
def main():
    print("Welcome To Fahrenheit to celsius convertor! :)")
    Fahren_Inp =input("Enter temprature in Fahrenheit: ")
    Fahren_Inp = float(Fahren_Inp)
    Celsius_Conv= (Fahren_Inp - 32) * 5.0/9.0
    print(f"Temperature: {Fahren_Inp}F = {Celsius_Conv}C")




if __name__ == '__main__':
    main()
```
