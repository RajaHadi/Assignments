# **Problem Statement**
Write a program that doubles each element in a list of numbers

# ***Solution***

```python
def main():
  numbers:list[int]=[1,2,3,4,5]
  for i in range(len(numbers)):
   elements_i=numbers[i]
   numbers[i]=elements_i * 2
  print(numbers)
if __name__ == '__main__':
    main()
```
